### 知识点：小端存储

这里有一个重要的知识点，由于IDA是反编译C语言，```s = ' fo`guci'``` 应该逆序处理

> 就是低位字节排放在内存的低地址端即该值的起始地址，高位字节排放在内存的高地址端。

````python
s = "fo`guci"
f2 =  s[::-1]
print(f2)
````

所以 `f2` 应该为 ```icug`of```

最后依据 `case 4uLL:` 和 `case 5uLL:` 计算过程，计算 `f2` ，然后将其与 `f1` 相加即可得到 `Flag`


````python
s = "fo`guci"

f1 = ""

# s 倒序
f2 =  s[::-1] 

flag = "GXY{do_not_"

for i in range(8):
    if i % 2 == 1:
        flag += chr(ord(f2[i]) - 2)
    else:
        flag += chr(ord(f2[i]) - 1)

print(flag)
````
